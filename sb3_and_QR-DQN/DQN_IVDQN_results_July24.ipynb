```python
def plot_data_max_last(DQN_data_0, DQN_data_8, DQN_data_15, IVDQN_data_0, IVDQN_data_8, IVDQN_data_15):

    mean_DQN_data = np.mean([DQN_data_0, DQN_data_8, DQN_data_15], axis=0)


    std_DQN_data = np.std([DQN_data_0, DQN_data_8, DQN_data_15], axis=0)


    mean_IVDQN_data = np.mean([IVDQN_data_0, IVDQN_data_8, IVDQN_data_15], axis=0)


    std_IVDQN_data = np.std([IVDQN_data_0, IVDQN_data_8, IVDQN_data_15], axis=0)
    
    print("Mean DQN return ", np.mean(mean_DQN_data))
    print("Mean IVDQN return ", np.mean(mean_IVDQN_data))
    print("Std DQN return ", np.std(mean_DQN_data))
    print("Std IVDQN return ", np.std(mean_IVDQN_data))
    
    print("Mean of last 10 retuns DQN ", np.mean(mean_DQN_data[-10:]))
    print("Mean of last 10 retuns IVDQN ", np.mean(mean_IVDQN_data[-10:]))
    print("Std of last 10 retuns DQN ", np.std(mean_DQN_data[-10:]))
    print("Std of last 10 retuns IVDQN ", np.std(mean_IVDQN_data[-10:]))
    

    plt.plot(mean_DQN_data, label='DQN')
    plt.fill_between(range(len(mean_DQN_data)), mean_DQN_data-std_DQN_data, mean_DQN_data+std_DQN_data, alpha=0.3)
    plt.plot(mean_IVDQN_data, label='IVDQN')
    plt.fill_between(range(len(mean_IVDQN_data)), mean_IVDQN_data-std_IVDQN_data, mean_IVDQN_data+std_IVDQN_data, alpha=0.3)
    plt.legend()
    plt.xlabel('Episodes')
    plt.ylabel('Average Episodic Reward')

    max_DQN_0 = np.max(DQN_data_0)
    max_DQN_8 = np.max(DQN_data_8)
    max_DQN_15 = np.max(DQN_data_15)

    max_IVDQN_0 = np.max(IVDQN_data_0)
    max_IVDQN_8 = np.max(IVDQN_data_8)
    max_IVDQN_15 = np.max(IVDQN_data_15)

    mean_max_DQN = np.mean([max_DQN_0, max_DQN_8, max_DQN_15])
    std_max_DQN = np.std([max_DQN_0, max_DQN_8, max_DQN_15])

    mean_max_IVDQN = np.mean([max_IVDQN_0, max_IVDQN_8, max_IVDQN_15])
    std_max_IVDQN = np.std([max_IVDQN_0, max_IVDQN_8, max_IVDQN_15])

    last_DQN_0 = DQN_data_0[-1]
    last_DQN_8 = DQN_data_8[-1]
    last_DQN_15 = DQN_data_15[-1]

    last_IVDQN_0 = IVDQN_data_0[-1]
    last_IVDQN_8 = IVDQN_data_8[-1]
    last_IVDQN_15 = IVDQN_data_15[-1]

    mean_last_DQN = np.mean([last_DQN_0, last_DQN_8, last_DQN_15])
    std_last_DQN = np.std([last_DQN_0, last_DQN_8, last_DQN_15])

    mean_last_IVDQN = np.mean([last_IVDQN_0, last_IVDQN_8, last_IVDQN_15])
    std_last_IVDQN = np.std([last_IVDQN_0, last_IVDQN_8, last_IVDQN_15])

    print('Max DQN:', mean_max_DQN, '+-', std_max_DQN)
    print('Max IVDQN:', mean_max_IVDQN, '+-', std_max_IVDQN)
    print('Last DQN:', mean_last_DQN, '+-', std_last_DQN)
    print('Last IVDQN:', mean_last_IVDQN, '+-', std_last_IVDQN)

    # return mean_max_DQN, std_max_DQN, mean_max_IVDQN, std_max_IVDQN, mean_last_DQN, std_last_DQN, mean_last_IVDQN, std_last_IVDQN

```
